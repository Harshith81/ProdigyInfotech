1. First we have to make a mark of the year in which a country has highest population like that for each country we have to make a graph
2. we have to plot a graph for the average population of each country

2. Plot a graph for country v/s average population across all years

```python
# This is the set of all country codes as an array
countryval=df['Country Code']
print(len(countryval))
countryval
```

```python
# This is the varaiation of population in the year 1960 (starting year) across various country's 
sns.boxplot(x=countryval,y=df[1960.0])
```

```python
# This is the varaiation of population in the year 2022 (Ending year) across various country's
sns.boxplot(data=df,x=countryval,y=df[2022.0])
```

```python
# This is the set of mean values of population of all countries over the range from 1960 to 2022
meanval=[]
for i in range(0,266):
    val=df.iloc[i]
    newval=val[4:].mean()
    meanval.append(newval)
print(len(meanval))
meanval
```

```python
# This is plot for country v/s average popular 
sns.boxplot(x=countryval,y=meanval)
```

```python
# This is the set for having all the years where the population is highest for a country
yearval=[]
for i in range(0,266):
    maxval=np.max(df.iloc[i,4:])
    for j in range(3,67):
        if df.iloc[i,j]==maxval:
            yearval.append(df.columns[j])
print(len(yearval))
yearval
```

```python
# This is the plot for country v/s Year where the population is highest.
sns.boxplot(x=countryval[:-1],y=yearval)
```
